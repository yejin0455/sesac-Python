```python
import pandas as pd
path = '/content/drive/MyDrive/module/train_titanic.csv'
data = pd.read_csv(path)
data
```

```python
names_values = ['None', 'Mr', 'Mrs', 'Miss', 'Master', 'Dr']
ndata = data['Name']
for ne in ndata:
    f, s = ne.split(', ')
    ses = s.split('.')
    s0 = ses[0]
    s0 = s0.replace(' ', '')
    check = False
    for i, ne2 in enumerate(names_values):
        if ne2 == s0:
            print(i)
            check = True
            break
        if check == False:
            print(0)
```

```python
datas = ["a", "b", "c"]
for i, d in enumerate(datas):
    print(i, d)
```
