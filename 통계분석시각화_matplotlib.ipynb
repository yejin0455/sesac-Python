# plot 함수

```python
import numpy as np
data = np.arange(10)  #y축 데이터
data
```

```python
plt.plot(data)  #도면에 추가, x축 데이터는 0부터 순차적으로 자동 부여
plt.show()  #시각화
```

```python
#y값만 인자로 전달, x축 데이터는 0부터 순차적으로 자동 부여
plt.plot([1, 4, 7, 9, 12, 15])
plt.show() 
```

```python
#제목 부여하기
plt.plot([1, 4, 7, 9, 12, 15])  #도면에 추가
plt.title("plt.plot([1, 4, 7, 9, 12, 15])")  #제목 달기
plt.show()  #시각화
```

```python
#x, y 전달
plt.plot([1, 2, 3, 4, 5, 7], [1, 4, 7, 9, 12, 15])
plt.title("plt.plot([1, 2, 3, 4, 5, 7], [1, 4, 7, 9, 12, 15])")
plt.show() 
```

'[marker][line][color]'

```python
#marker 사용
plt.plot([1, 2, 3, 4, 5, 7], [1, 2, 9, 7, 12, 15], '.')  #.으로 표시
plt.title("plt.plot([1, 2, 3, 4, 5, 7], [1, 4, 7, 9, 12, 15]).'.'")
plt.show() 
```

```python
#marker 사용
plt.plot([1, 2, 3, 4, 5, 7], [1, 2, 9, 7, 12, 15], 'o')  #원형으로 표시
plt.title("plt.plot([1, 2, 3, 4, 5, 7], [1, 4, 7, 9, 12, 15]).'o'")
plt.show() 
```

```python
#marker, line 사용
plt.plot([1, 2, 3, 4, 5, 7], [1, 2, 9, 7, 12, 15], 'o-')  #원형 솔리드로 표시
plt.title("plt.plot([1, 2, 3, 4, 5, 7], [1, 4, 7, 9, 12, 15]).'o-'")
plt.show() 
```

```python
#marker, line 사용
plt.plot([1, 2, 3, 4, 5, 7], [1, 2, 9, 7, 12, 15], 'o:')  #원형 닷으로 표시
plt.title("plt.plot([1, 2, 3, 4, 5, 7], [1, 4, 7, 9, 12, 15]).'o:'")
plt.show() 
```

```python
#marker, line, color 사용
plt.plot([1, 2, 3, 4, 5, 7], [1, 2, 9, 7, 12, 15], 'o-.r')  #원형 대쉬닷 빨간색으로 표시
plt.title("plt.plot([1, 2, 3, 4, 5, 7], [1, 4, 7, 9, 12, 15]).'o-.r'")
plt.show() 
```

```python
#kwargs의 color 사용
plt.plot([1, 2, 3, 4, 5, 7], [1, 2, 9, 7, 12, 15], 'o--', color='darkgreen')  #원형 대쉬라인 kwargs 사용해서 표시
plt.title("plt.plot([1, 2, 3, 4, 5, 7], [1, 4, 7, 9, 12, 15]).'o--',color=darkgreen")
plt.show() 
```

```python
#kwargs의 color RGB 사용 
plt.plot([1, 2, 3, 4, 5, 7], [1, 2, 9, 7, 12, 15], 'o--', color='#FF00FF')  #원형 대쉬라인 RGB 사용해서 표시
plt.title("plt.plot([1, 2, 3, 4, 5, 7], [1, 4, 7, 9, 12, 15]).'o--',color=#FF00FF")
plt.show() 
```

```python
#kwargs의 linewidth 사용
plt.plot([1, 2, 3, 4, 5, 7], [1, 2, 9, 7, 12, 15], 'o--', color='darkgreen', linewidth=5)  #원형 대쉬라인 kwargs 사용해서 표시
plt.title("plt.plot([1, 2, 3, 4, 5, 7], [1, 4, 7, 9, 12, 15]).'o--',color=darkgreen, linewidth=5")
plt.show() 
```

```python
#데이터를 두 개 이상 전달하기
plt.plot([1, 2, 3, 4, 5, 7], [1, 2, 9, 7, 12, 15], 'bo', [1, 2, 3, 4, 5, 7], [8, 9, 10, 15, 10, 11], 'r^')
plt.title("plt.plot([1, 2, 3, 4, 5, 7], [1, 2, 9, 7, 12, 15], 'bo', [1, 2, 3, 4, 5, 7], [8, 9, 10, 15, 10, 11], 'r^'")
plt.show() 
```

```python
#데이터를 두 개 이상 전달하기
plt.plot([1, 2, 3, 4, 5, 7], [1, 2, 9, 7, 12, 15], 'bo')
plt.plot([1, 2, 3, 4, 5, 7], [8, 9, 10, 15, 10, 11], 'r^')
plt.title("plt.plot([1, 2, 3, 4, 5, 7], [1, 2, 9, 7, 12, 15], 'bo', [1, 2, 3, 4, 5, 7], [8, 9, 10, 15, 10, 11], 'r^'")
plt.show() 
```

```python
#축 이름 지정하기
plt.plot([1, 2, 3, 4, 5], [90, 80, 85, 77, 97], 'bo')
plt.plot([1, 2, 3, 4, 5], [95, 81, 87, 78, 94], 'r.')
plt.axis([0, 6, 0, 100])
plt.xlabel('number')
plt.ylabel('scores')
plt.title("scores")
plt.show() 
```

```python
#축 이름 지정하기
plt.plot([1, 2, 3, 4, 5], [90, 80, 85, 77, 97], 'bo', label='class 1')
plt.plot([1, 2, 3, 4, 5], [95, 81, 87, 78, 94], 'r.', label='class 2')
plt.axis([0, 6, 0, 100])
plt.xlabel('number')
plt.ylabel('scores')
plt.title("scores")
plt.legend() #label를 표시하세요
plt.show() 
```

```python
#sub 도표 사용하기
plt.subplot(221)  #2x2 도표 중에 1
plt.title("221")
plt.plot([1, 2, 3, 4], 'ro-')

plt.subplot(222)  #2x2 도표 중에 2
plt.title("222")
plt.plot([4, 3, 2, 1], 'b.:')

plt.subplot(223)  #2x2 도표 중에 3
plt.title("223")
plt.plot([1, 3, 2, 4], 'c^-.')

plt.subplot(224)  #2x2 도표 중에 4
plt.title("224")
plt.plot([2, 4, 1, 3], 'k.--')

plt.show()
```

```python
#y축에 표현할 값을 지정 + 디폴트 색상 확인
ys_str = []
for i in range(1, 31):
    plt.plot([i, i])
    if i % 3 == 0:
        ys_str.append(str(i))
    else:
        ys_str.append(" ")
plt.title("default color, set yticks")
plt.yticks(range(1, 31), ys_str)
plt.show()
```

```python
#배경색
plt.plot([1, 1], 'ro-')
ax = plt.gca()
ax.set_facecolor('violet')
plt.title("set background color")
plt.show()
```

```python
#data 사용하기
mdic = dict()
mdic['hour'] = [0, 1, 2, 3, 4, 5]
mdic['distance'] = [0, 5, 20, 45, 70, 95]

plt.plot('hour', 'distance', data=mdic)
plt.title("using data")
plt.savefig('demo.png')
plt.show()
```

```python
#에러 값 표현하기
colors = ['red','green','blue']
names = ['a','b','c']
income = [200,230,300]
err = [2, 3, 10]
plt.bar(names,income,color=colors,yerr=err)
plt.title("에러 값 표현")
plt.show()
```

```python
#에러 값 표현하기
colors = ['red','green','blue']
names = ['a','b','c']
income = [200,230,300]
err = [2, 3, 10]
plt.barh(names,income,color=colors,xerr=err)
plt.title("에러 값 표현")
plt.show()
```

```python
#에러 값 표현하기
colors = ['red','green','blue']
ecolors = ['g', 'b', 'r']
names = ['a','b','c']
income = [200,230,300]
err = [2, 3, 10]
plt.bar(names,income,color=colors,yerr=err, ecolor=ecolors)
plt.title("에러 값 표현")
plt.show()
```

```python
#에러 값 표현하기, 캡 달기
colors = ['red','green','blue']
ecolors = ['green', 'blue', 'red']
names = ['a','b','c']
income = [200,230,300]
err = [2, 3, 10]
plt.bar(names,income,color=colors,yerr=err, ecolor='black', capsize=4.0)  #캡 달면 에러마다 색상부여 못함
plt.title("에러 값 표현, 캡 달기")
plt.show()
```

# 파이그래프



```python
locales = ['서울', '강원', '경북', '충남', '기타']
incomes = [2300, 1200, 1800, 800, 400]
plt.pie(incomes, labels=locales)
plt.title("지역별 수입")
plt.show()
```

```python
#돌출시키기
locales = ['서울', '강원', '경북', '충남', '기타']
incomes = [2300, 1200, 1800, 800, 400]
exs = [0, 0.1, 0, 0.9, 0]
plt.pie(incomes, labels=locales, explode=exs)
plt.title("지역별 수입-돌출시키기")
plt.show()
```

```python
#퍼센트 표시하기
locales = ['서울', '강원', '경북', '충남', '기타']
incomes = [2300, 1200, 1800, 800, 400]
exs = [0, 0.1, 0, 0, 0]
plt.pie(incomes, labels=locales, explode=exs, autopct='%.2f%%')
plt.title("지역별 수입-퍼센트표시하기")
plt.show()
```

```python
#퍼센트 표시할 위치 지정하기
locales = ['서울', '강원', '경북', '충남', '기타']
incomes = [2300, 1200, 1800, 800, 400]
exs = [0, 0.1, 0, 0, 0]
plt.pie(incomes, labels=locales, explode=exs, autopct='%.2f%%', pctdistance=0.8)
plt.title("지역별 수입-퍼센트표시할 위치 지정하기")
plt.show()
```

```python
#그림자 효과
locales = ['서울', '강원', '경북', '충남', '기타']
incomes = [2300, 1200, 1800, 800, 400]
exs = [0, 0.1, 0, 0, 0]
plt.pie(incomes, labels=locales, explode=exs, autopct='%.2f%%', pctdistance=0.8, shadow=True)
plt.title("지역별 수입-그림자표시하기")
plt.show()
```

```python
#그림자 효과
locales = ['서울', '강원', '경북', '충남', '기타']
incomes = [2300, 1200, 1800, 800, 400]
exs = [0, 0.1, 0, 0, 0]
plt.pie(incomes, labels=locales, explode=exs, autopct='%.2f%%', pctdistance=0.8, labeldistance=1.2)
plt.title("지역별 수입-label 표시할 위치 지정하기")
plt.show()
```

```python
#시계 방향으로 각도 계산하기
locales = ['서울', '강원', '경북', '충남', '기타']
incomes = [2300, 1200, 1800, 800, 400]
exs = [0, 0.1, 0, 0, 0]
plt.pie(incomes, labels=locales, explode=exs, autopct='%.2f%%', startangle=90, counterclock=False)
plt.title("지역별 수입-시계 방향으로 각도 계산하기")
plt.show()
```
